```python
# closest location Cambridge Maryland
# https://tidesandcurrents.noaa.gov/waterlevels.html?id=8571892

# Link for CSV:
# https://api.tidesandcurrents.noaa.gov/api/prod/datagetter
#  ?product=water_level&application=NOS.COOPS.TAC.WL
#  &begin_date=20250201&end_date=20250202&datum=MLLW
#  &station=8571892&time_zone=GMT&units=english&format=csv

import pandas as pd
import sqlite3
import plotly.express as px
```

```python
conn = sqlite3.connect('water_levels.db')
scur = conn.cursor()
```

```python
scur = conn.cursor()
scur.execute(f'''
select *
from daily
limit 10''')

for col in scur.description:
    print(col[0])

for row in scur.fetchall():
    print(row)
```

```python

scur.execute(f'''
select Station, date(report_date,'start of year') as Year,
avg(avg_water_level),
avg(max_water_level), 
avg(min_water_level), 
max(max_water_level),
sum(record_count), count(*)
from daily

group by 1,2
order by 1,2
''')

```

```python
# data = cursor.fetchall()

# df = pd.DataFrame(data, columns=['column1', 'column2'])

df = pd.DataFrame(scur.fetchall(), columns=list(map(lambda x: x[0], scur.description))) 

```

```python
df
```

```python
list(df.columns)
```

```python
cambridge = df[df['Station'] == '8571892']
annapolis = df[df['Station'] == '8575512']
washington = df[df['Station'] == '8594900']
```

```python
 import plotly.express as px
fig = px.scatter(cambridge, x='Year', y='avg(avg_water_level)', title='Cambridge Station')
fig.add_scatter(x=cambridge['Year'], y=df['avg(max_water_level)'], name="high water levels")
fig.add_scatter(x=cambridge['Year'], y=df['avg(min_water_level)'], name="low water levels")
fig.add_scatter(x=cambridge['Year'], y=df['max(max_water_level)'], 
                name="highest water levels",
                marker=dict(color="Red") )

fig.show()
```

```python
 import plotly.express as px
fig = px.scatter(washington, x='Year', y='avg(avg_water_level)', title='Washington Station')
fig.add_scatter(x=washington['Year'], y=df['avg(max_water_level)'], name="high water levels")
fig.add_scatter(x=washington['Year'], y=df['avg(min_water_level)'], name="low water levels")
fig.add_scatter(x=washington['Year'], y=df['max(max_water_level)'], name="highest water levels")
fig.add_scatter(x=washington['Year'], y=df['max(max_water_level)'], 
                name="highest water levels",
                marker=dict(color="Red") )

fig.show()
```
